## Path A — Build a free-form analytics query

Pick a question about the Olist dataset that genuinely interests you and answer it with SQL. Use any combination of the techniques you've learned. A few ideas to get you started:

### Question ideas

| Question | Concepts you'd use |
|---|---|
| Which sellers have the highest review scores AND the most orders? | JOIN, GROUP BY, aggregation |
| Do customers who pay in instalments spend more on average? | GROUP BY, AVG, CASE WHEN |
| What's the average delivery time by product category? | JOIN, DATEDIFF, GROUP BY |
| Which cities have the highest late delivery rate? | JOIN, CASE WHEN, GROUP BY, HAVING |
| Are Champions (from RFM) more likely to leave positive reviews? | CTEs, RFM logic, JOIN to reviews |
| What's the repeat purchase rate by cohort month? | Cohort analysis, retention logic |
| Which product categories have the most canceled orders? | JOIN, GROUP BY, conditional aggregation |
| Do orders placed on weekends take longer to deliver? | DAYOFWEEK, DATEDIFF, GROUP BY |

### Tips for a good free-form query

- **Start with a question**, not a query. Write the question as a comment at the top.
- **Break it into CTEs.** Name each step clearly — it makes debugging much easier.
- **Sanity check your results.** Do the numbers make sense? Spot-check a few rows.
- **Add comments.** Future you (and your teammates) will thank you.

## Path B — Build a dashboard

Take the queries you've written throughout the challenge and turn them into a live dashboard in the tool of your choice. You already have everything you need — the hard work is done.

### Suggested dashboard structure

A good e-commerce analytics dashboard typically has 3–4 pages:

**Page 1 — Revenue Overview**
- Monthly revenue trend (line chart) — your Week 6 query
- Month-over-month growth rate (KPI card)
- Revenue by product category (bar chart) — your Week 12 query

**Page 2 — Customer Retention**
- Cohort retention heatmap — your Week 7 query
- RFM segment breakdown (donut or bar chart) — your Week 9 query
- Customer count by segment over time

**Page 3 — Operations**
- Order funnel conversion rates (funnel chart) — your Week 8 query
- Average delivery time by state (map or bar chart)
- Late delivery rate trend over time

**Page 4 — Free choice**
- Whatever question from Path A that you found most interesting

### Connecting your dashboard tool to Snowflake

Most BI tools connect to Snowflake the same way. You'll need:

| Setting | Value |
|---|---|
| Account | Your Snowflake account identifier |
| Warehouse | `ANALYTICS_TEAM_WH` |
| Database | `ANALYTICS_TEAM_DB` |
| Schema | `SQL_CHALLENGE` |
| Role | `ANALYTICS_TEAM_ROLE` |

**Sigma Computing** — Administration → Connections → Add Connection → Snowflake

**Tableau** — Connect → To a Server → Snowflake → enter credentials above

**Looker Studio** — Add data → Snowflake connector → enter credentials above

**Power BI** — Get Data → Snowflake → enter server and warehouse

> **Tip:** In your dashboard tool, paste your SQL queries directly as custom SQL data sources rather than dragging tables. Your CTEs will work exactly the same way.
