## What you're practicing this week

Last week you ranked rows. This week you look forwards and backwards across them. `LAG` and `LEAD` let you compare each row to the row before or after it — essential for period-over-period comparisons. Running totals accumulate a value row by row over time.

## New concepts this week

```sql
-- LAG: look at the previous row's value
SELECT
    order_month,
    order_count,
    LAG(order_count, 1) OVER (ORDER BY order_month) AS prev_month_count
FROM monthly_orders;
-- The '1' means 1 row back. LAG(col, 2) would look 2 rows back.
-- The first row returns NULL because there's no row before it.
```

```sql
-- LEAD: look at the next row's value
SELECT
    order_month,
    order_count,
    LEAD(order_count, 1) OVER (ORDER BY order_month) AS next_month_count
FROM monthly_orders;
-- The last row returns NULL because there's no row after it.
```

```sql
-- Running total: accumulate values from first row to current
SELECT
    order_month,
    order_count,
    SUM(order_count) OVER (ORDER BY order_month
        ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW) AS running_total
FROM monthly_orders;
```

```sql
-- 3-month moving average
SELECT
    order_month,
    order_count,
    ROUND(AVG(order_count) OVER (ORDER BY order_month
        ROWS BETWEEN 2 PRECEDING AND CURRENT ROW), 1) AS moving_avg_3m
FROM monthly_orders;
```

````sql
-- Example 2: LAG to show previous month alongside current
WITH monthly AS (
    SELECT
        DATE_TRUNC('month', order_purchase_timestamp) AS order_month,
        COUNT(*) AS order_count
    FROM ANALYTICS_TEAM_DB.SQL_CHALLENGE.OLIST_ORDERS
    GROUP BY order_month
)
SELECT
    order_month,
    order_count,
    LAG(order_count, 1) OVER (ORDER BY order_month) AS prev_month_count
FROM monthly
ORDER BY order_month;
````

````sql
-- Example 3: running total
WITH monthly AS (
    SELECT
        DATE_TRUNC('month', order_purchase_timestamp) AS order_month,
        COUNT(*) AS order_count
    FROM ANALYTICS_TEAM_DB.SQL_CHALLENGE.OLIST_ORDERS
    GROUP BY order_month
)
SELECT
    order_month,
    order_count,
    SUM(order_count) OVER (ORDER BY order_month
        ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW) AS running_total
FROM monthly
ORDER BY order_month;
````
