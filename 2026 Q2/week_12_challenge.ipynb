## The translation table

This week you finally use it. Product categories are in Portuguese in `OLIST_PRODUCTS` — this table maps them to English.

```sql
SELECT * FROM ANALYTICS_TEAM_DB.SQL_CHALLENGE.PRODUCT_CATEGORY_NAME_TRANSLATION LIMIT 10;
```

```sql
-- Example: how to join to get English category names
-- Use LEFT JOIN so untranslated categories aren't silently dropped
SELECT
    p.product_category_name,
    COALESCE(t.product_category_name_english, p.product_category_name) AS category_english
FROM ANALYTICS_TEAM_DB.SQL_CHALLENGE.OLIST_PRODUCTS p
LEFT JOIN ANALYTICS_TEAM_DB.SQL_CHALLENGE.PRODUCT_CATEGORY_NAME_TRANSLATION t
    ON p.product_category_name = t.product_category_name
LIMIT 10;
```
