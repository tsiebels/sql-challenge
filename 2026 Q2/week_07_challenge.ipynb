## What you're practicing this week

Cohort analysis groups customers by when they first appeared and tracks their behaviour over time. It answers questions like: *of the customers who first ordered in January, how many came back in February?* It's one of the most important techniques in e-commerce analytics.

## New concepts this week

### Step 1: Assign every customer a cohort
A cohort is defined by the month of a customer's first ever order:

```sql
-- Find each customer's first order month
WITH cohorts AS (
    SELECT
        customer_id,
        DATE_TRUNC('month', MIN(order_purchase_timestamp)) AS cohort_month
    FROM OLIST_ORDERS
    GROUP BY customer_id
)
SELECT * FROM cohorts LIMIT 5;
```

### Step 2: Calculate months since first order
`DATEDIFF` measures the gap between two dates:

```sql
-- months_since_first_order = 0 means they ordered in their first month
DATEDIFF('month', cohort_month, DATE_TRUNC('month', order_purchase_timestamp))
```

### Step 3: Calculate retention
```sql
-- For each cohort + period, count distinct active customers
-- Divide by cohort size to get the retention rate
SELECT
    cohort_month,
    months_since_first_order,
    COUNT(DISTINCT customer_id) AS active_customers,
    ROUND(COUNT(DISTINCT customer_id) * 100.0 / cohort_size, 1) AS retention_rate
```

## Try the examples

````sql
-- Example 1: find each customer's cohort month
WITH cohorts AS (
    SELECT
        customer_id,
        DATE_TRUNC('month', MIN(order_purchase_timestamp)) AS cohort_month
    FROM ANALYTICS_TEAM_DB.SQL_CHALLENGE.OLIST_ORDERS
    GROUP BY customer_id
)
SELECT cohort_month, COUNT(*) AS cohort_size
FROM cohorts
GROUP BY cohort_month
ORDER BY cohort_month;
````

````sql
-- Example 2: join cohorts back to all orders and calculate months since first order
WITH cohorts AS (
    SELECT customer_id,
           DATE_TRUNC('month', MIN(order_purchase_timestamp)) AS cohort_month
    FROM ANALYTICS_TEAM_DB.SQL_CHALLENGE.OLIST_ORDERS
    GROUP BY customer_id
)
SELECT
    o.customer_id,
    c.cohort_month,
    DATE_TRUNC('month', o.order_purchase_timestamp) AS order_month,
    DATEDIFF('month', c.cohort_month,
        DATE_TRUNC('month', o.order_purchase_timestamp)) AS months_since_first_order
FROM ANALYTICS_TEAM_DB.SQL_CHALLENGE.OLIST_ORDERS o
JOIN cohorts c ON o.customer_id = c.customer_id
LIMIT 10;
````
