## What you're practicing this week

Funnel analysis measures how many customers make it through each stage of a process. You'll also meet two new tables this week — products and sellers — and learn conditional aggregation, which lets you count subsets of rows in a single query without multiple GROUP BYs.

## New concepts this week

### Conditional aggregation
Instead of running separate queries for each status, count everything in one pass using `CASE WHEN` inside `COUNT`:

```sql
SELECT
    COUNT(*) AS total_orders,
    COUNT(CASE WHEN order_status = 'approved'  THEN 1 END) AS approved,
    COUNT(CASE WHEN order_status = 'shipped'   THEN 1 END) AS shipped,
    COUNT(CASE WHEN order_status = 'delivered' THEN 1 END) AS delivered
FROM OLIST_ORDERS;
```

### COUNT(DISTINCT ...)
When rows can repeat per customer or order, use `COUNT(DISTINCT col)` to count unique values:

```sql
-- Number of distinct orders each seller appears in
SELECT seller_id, COUNT(DISTINCT order_id) AS order_count
FROM OLIST_ORDER_ITEMS
GROUP BY seller_id;
```

### The Olist order funnel
```
Placed → Approved → Invoiced → Shipped → Delivered
         ↓                              ↓
      Canceled                        Unavailable
```

````sql
-- Example 2: COUNT vs COUNT(DISTINCT) — see the difference
SELECT
    COUNT(order_id)          AS total_rows,
    COUNT(DISTINCT order_id) AS unique_orders
FROM ANALYTICS_TEAM_DB.SQL_CHALLENGE.OLIST_ORDER_ITEMS;
````
