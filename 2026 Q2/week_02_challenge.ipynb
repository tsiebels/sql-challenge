## What you're practicing this week

Summarising rows instead of just filtering them. Aggregations are the heartbeat of BI work — nearly every dashboard metric, KPI, and stakeholder report has a `GROUP BY` underneath it. This week stays on the orders table so you can focus entirely on the new concepts without worrying about joins yet.

## New concepts this week

| Concept | What it does |
|---|---|
| `COUNT(*)` | Counts all rows in a group |
| `COUNT(column)` | Counts non-NULL values only — different from `COUNT(*)` |
| `SUM(column)` | Adds up values in a group |
| `AVG(column)` | Averages values in a group |
| `GROUP BY` | Splits rows into groups before aggregating |
| `HAVING` | Filters on aggregated results — like `WHERE` but runs after grouping |
| `DATE_TRUNC` | Rounds a timestamp down to the start of a period (month, year, etc.) |

A quick reminder of the order SQL runs in — this trips a lot of people up:

```
FROM → WHERE → GROUP BY → HAVING → SELECT → ORDER BY
```

This means you **cannot** use an alias you defined in `SELECT` inside a `WHERE` clause — the `WHERE` runs before `SELECT`. You **can** use it in `ORDER BY` though, because that runs last.

## Try the examples

Run these before you start the tasks to see how aggregations work.

````sql
-- Example 1: COUNT(*) — count all rows
SELECT COUNT(*) AS total_orders
FROM ANALYTICS_TEAM_DB.SQL_CHALLENGE.OLIST_ORDERS;
````

````sql
-- Example 2: GROUP BY — count rows per group
SELECT order_status, COUNT(*) AS order_count
FROM ANALYTICS_TEAM_DB.SQL_CHALLENGE.OLIST_ORDERS
GROUP BY order_status
ORDER BY order_count DESC;
````

````sql
-- Example 3: SUM and AVG
SELECT
    COUNT(*)                    AS total_orders,
    COUNT(order_delivered_customer_date) AS delivered_orders,  -- skips NULLs
    COUNT(*) - COUNT(order_delivered_customer_date) AS undelivered -- NULL count
FROM ANALYTICS_TEAM_DB.SQL_CHALLENGE.OLIST_ORDERS;
````

````sql
-- Example 4: DATE_TRUNC — group by month
-- Rounds a timestamp down to the start of the period
SELECT
    DATE_TRUNC('month', order_purchase_timestamp) AS order_month,
    COUNT(*) AS order_count
FROM ANALYTICS_TEAM_DB.SQL_CHALLENGE.OLIST_ORDERS
GROUP BY order_month
ORDER BY order_month;
````

````sql
-- Example 5: HAVING — filter after aggregation
-- WHERE filters rows before grouping; HAVING filters groups after
SELECT order_status, COUNT(*) AS order_count
FROM ANALYTICS_TEAM_DB.SQL_CHALLENGE.OLIST_ORDERS
GROUP BY order_status
HAVING COUNT(*) > 500
ORDER BY order_count DESC;
````

````sql
-- Example 6: CASE WHEN — create a label column
SELECT
    order_id,
    CASE
        WHEN order_status = 'delivered' THEN 'Complete'
        WHEN order_status = 'canceled'  THEN 'Canceled'
        ELSE 'In Progress'
    END AS order_category
FROM ANALYTICS_TEAM_DB.SQL_CHALLENGE.OLIST_ORDERS
LIMIT 10;
````
