## What you're practicing this week

Connecting two tables together. Real BI work almost always requires joining multiple tables — a single table rarely contains everything you need. This week you'll link orders to customers and start to understand the important difference between `INNER JOIN` and `LEFT JOIN`.

## New concepts this week

| Concept | What it does |
|---|---|
| `INNER JOIN` | Returns only rows that have a match in **both** tables |
| `LEFT JOIN` | Returns **all** rows from the left table, with NULLs where there's no match on the right |
| `ON` | Defines which columns connect the two tables |
| Table aliases | Shorthand names like `o` and `c` to avoid typing full table names |

**The two tables this week:**

`ANALYTICS_TEAM_DB.SQL_CHALLENGE.OLIST_ORDERS` and `ANALYTICS_TEAM_DB.SQL_CHALLENGE.OLIST_CUSTOMERS` share a `customer_id` column — that's your join key.

```
OLIST_ORDERS                    OLIST_CUSTOMERS
─────────────────────────────   ─────────────────────────────────
order_id                        customer_id  ← join key
customer_id  ← join key         customer_city
order_status                    customer_state
order_purchase_timestamp        customer_zip_code_prefix
...                             ...
```

Get into the habit of **aliasing your tables** as soon as you start joining. It keeps your queries readable and avoids ambiguity when two tables share a column name:

```sql
FROM ANALYTICS_TEAM_DB.SQL_CHALLENGE.OLIST_ORDERS o
INNER JOIN ANALYTICS_TEAM_DB.SQL_CHALLENGE.OLIST_CUSTOMERS c ON o.customer_id = c.customer_id
```

## Try the examples

Run these to see INNER JOIN and LEFT JOIN in action before tackling the tasks.

````sql
-- Example 1: INNER JOIN — only rows that match in both tables
-- customer_id must exist in both OLIST_ORDERS and OLIST_CUSTOMERS
SELECT
    o.order_id,
    o.order_status,
    c.customer_city,
    c.customer_state
FROM ANALYTICS_TEAM_DB.SQL_CHALLENGE.OLIST_ORDERS o
INNER JOIN ANALYTICS_TEAM_DB.SQL_CHALLENGE.OLIST_CUSTOMERS c ON o.customer_id = c.customer_id
LIMIT 10;
````

````sql
-- Example 2: LEFT JOIN — all rows from the left table,
-- NULLs on the right where there is no match
SELECT
    o.order_id,
    o.order_status,
    c.customer_city  -- will be NULL if no matching customer
FROM ANALYTICS_TEAM_DB.SQL_CHALLENGE.OLIST_ORDERS o
LEFT JOIN ANALYTICS_TEAM_DB.SQL_CHALLENGE.OLIST_CUSTOMERS c ON o.customer_id = c.customer_id
LIMIT 10;
````

````sql
-- Example 3: JOIN + GROUP BY — the most common BI pattern
-- Connect tables first, then aggregate the combined result
SELECT
    c.customer_state,
    COUNT(*) AS order_count
FROM ANALYTICS_TEAM_DB.SQL_CHALLENGE.OLIST_ORDERS o
INNER JOIN ANALYTICS_TEAM_DB.SQL_CHALLENGE.OLIST_CUSTOMERS c ON o.customer_id = c.customer_id
GROUP BY c.customer_state
ORDER BY order_count DESC
LIMIT 10;
````

````sql
-- Example 4: finding missing rows with LEFT JOIN + IS NULL
-- This pattern finds rows in the left table with no match on the right
SELECT c.customer_id, c.customer_city
FROM ANALYTICS_TEAM_DB.SQL_CHALLENGE.OLIST_CUSTOMERS c
LEFT JOIN ANALYTICS_TEAM_DB.SQL_CHALLENGE.OLIST_ORDERS o ON c.customer_id = o.customer_id
WHERE o.order_id IS NULL  -- no matching order = customer never ordered
LIMIT 10;
````

## A note on JOIN types

You've now seen `INNER JOIN` and `LEFT JOIN`. There are others — `RIGHT JOIN`, `FULL OUTER JOIN`, `CROSS JOIN` — but in practice, `INNER` and `LEFT` cover the vast majority of real-world BI queries. If you find yourself reaching for a `RIGHT JOIN`, it almost always means you should swap the table order and use a `LEFT JOIN` instead — it's easier to reason about.
