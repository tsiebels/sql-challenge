## New concepts this week

### Subquery
A query nested inside another query. You can put a subquery in the `FROM` clause (treating it like a table) or in the `WHERE` clause (to filter against a calculated value).

```sql
-- Subquery in WHERE: find orders above the average payment
SELECT order_id, payment_value
FROM OLIST_ORDER_PAYMENTS
WHERE payment_value > (
    SELECT AVG(payment_value)
    FROM OLIST_ORDER_PAYMENTS
);
```

```sql
-- Subquery in FROM: treat a grouped result as a table
SELECT customer_id, order_count
FROM (
    SELECT customer_id, COUNT(*) AS order_count
    FROM OLIST_ORDERS
    GROUP BY customer_id
) sub
WHERE order_count > 1;
```

### CTE (Common Table Expression)
A CTE gives a name to a query result so you can refer to it below. Use `WITH` at the top. CTEs are almost always cleaner than subqueries — prefer them.

```sql
-- Same logic as above, rewritten as a CTE
WITH order_counts AS (
    SELECT customer_id, COUNT(*) AS order_count
    FROM OLIST_ORDERS
    GROUP BY customer_id
)
SELECT customer_id, order_count
FROM order_counts
WHERE order_count > 1;
```

You can chain multiple CTEs by separating them with a comma:

```sql
WITH step_one AS (
    SELECT ...
),
step_two AS (
    SELECT ...
    FROM step_one  -- reference the first CTE
)
SELECT * FROM step_two;
```

## Try the examples

Run these to see subqueries and CTEs in action before you start the tasks.

````sql
-- Example 1: subquery in WHERE
-- Find orders where payment is above average
SELECT order_id, payment_value
FROM ANALYTICS_TEAM_DB.SQL_CHALLENGE.OLIST_ORDER_PAYMENTS
WHERE payment_value > (
    SELECT AVG(payment_value) FROM ANALYTICS_TEAM_DB.SQL_CHALLENGE.OLIST_ORDER_PAYMENTS
)
ORDER BY payment_value DESC
LIMIT 10;
````

````sql
-- Example 2: CTE version of the same logic
WITH avg_payment AS (
    SELECT AVG(payment_value) AS avg_val
    FROM ANALYTICS_TEAM_DB.SQL_CHALLENGE.OLIST_ORDER_PAYMENTS
)
SELECT p.order_id, p.payment_value
FROM ANALYTICS_TEAM_DB.SQL_CHALLENGE.OLIST_ORDER_PAYMENTS p
JOIN avg_payment a ON p.payment_value > a.avg_val
ORDER BY p.payment_value DESC
LIMIT 10;
````

````sql
-- Example 3: two CTEs chained together
WITH payment_totals AS (
    SELECT order_id, SUM(payment_value) AS total_payment
    FROM ANALYTICS_TEAM_DB.SQL_CHALLENGE.OLIST_ORDER_PAYMENTS
    GROUP BY order_id
),
orders_with_payment AS (
    SELECT o.order_id, o.order_status, p.total_payment
    FROM ANALYTICS_TEAM_DB.SQL_CHALLENGE.OLIST_ORDERS o
    JOIN payment_totals p ON o.order_id = p.order_id
)
SELECT * FROM orders_with_payment
ORDER BY total_payment DESC
LIMIT 10;
````
