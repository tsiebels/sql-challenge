## What you're practicing this week

## The dataset

All challenges use the **Olist Brazilian E-Commerce** dataset loaded into the `OLIST` schema in our shared Snowflake database. Olist is a Brazilian marketplace platform with ~100,000 real orders placed between 2016 and 2018.

This week you'll work exclusively with `ANALYTICS_TEAM_DB.SQL_CHALLENGE.OLIST_ORDERS`.

| Column | Type | What it contains |
|---|---|---|
| `order_id` | VARCHAR | Unique order identifier |
| `customer_id` | VARCHAR | Links to the customers table |
| `order_status` | VARCHAR | `'delivered'`, `'shipped'`, `'canceled'`, `'invoiced'`, `'processing'` |
| `order_purchase_timestamp` | TIMESTAMP | When the customer placed the order |
| `order_delivered_customer_date` | TIMESTAMP | Actual delivery date — can be NULL |
| `order_estimated_delivery_date` | TIMESTAMP | The promised delivery date |

## Try the examples

Run these before you start the tasks — they show you exactly how SELECT, WHERE, and ORDER BY work on real data.

```sql
-- Example 1: SELECT — choose specific columns
-- Instead of SELECT * (all columns), name the ones you want
SELECT order_id, order_status, order_purchase_timestamp
FROM ANALYTICS_TEAM_DB.SQL_CHALLENGE.OLIST_ORDERS
LIMIT 10;
```

```sql
-- Example 2: WHERE — filter rows by a condition
SELECT order_id, order_status, order_purchase_timestamp
FROM ANALYTICS_TEAM_DB.SQL_CHALLENGE.OLIST_ORDERS
WHERE order_status = 'shipped'
LIMIT 10;
```

```sql
-- Example 3: AND / OR — combine multiple conditions
SELECT order_id, order_status, order_purchase_timestamp
FROM ANALYTICS_TEAM_DB.SQL_CHALLENGE.OLIST_ORDERS
WHERE order_status = 'shipped'
  AND order_purchase_timestamp >= '2018-01-01'
LIMIT 10;
```

```sql
-- Example 4: ORDER BY — sort results
-- ASC = smallest/oldest first, DESC = largest/newest first
SELECT order_id, order_status, order_purchase_timestamp
FROM ANALYTICS_TEAM_DB.SQL_CHALLENGE.OLIST_ORDERS
WHERE order_status = 'delivered'
ORDER BY order_purchase_timestamp DESC
LIMIT 10;
```

```sql
-- Example 5: IN — match multiple values cleanly
SELECT order_id, order_status
FROM ANALYTICS_TEAM_DB.SQL_CHALLENGE.OLIST_ORDERS
WHERE order_status IN ('shipped', 'delivered')
LIMIT 10;
```

```sql
-- Example 6: IS NULL — find missing values
-- You cannot use = NULL in SQL, it will never match anything
SELECT order_id, order_status, order_delivered_customer_date
FROM ANALYTICS_TEAM_DB.SQL_CHALLENGE.OLIST_ORDERS
WHERE order_delivered_customer_date IS NULL
LIMIT 10;
```
